```python
from faker import Faker
from collections import OrderedDict
import pandas as pd
import numpy as np
```

```python
fake = Faker()
```

```python
def create_data(x): 

    b_user ={} 
    for i in range(0, x): 
        b_user[i] = {}
        b_user[i]['Width'] = fake.pyfloat(right_digits=1, positive=True, max_value=1000)
        b_user[i]['Height'] = fake.pyfloat(right_digits=4, positive=True, max_value=500)
        b_user[i]['Amount'] = fake.pyint()
        b_user[i]['Color'] = fake.pyfloat(right_digits=2, min_value=0, max_value=1)
        b_user[i]['Avg_speed'] = fake.pyint(min_value=60, max_value=365)
        b_user[i]['Weight'] = fake.pyfloat(right_digits=4, positive=True, max_value=150)
        b_user[i]['Temperature(max)'] = fake.pyfloat(right_digits=2, min_value=0, max_value=999)
        b_user[i]['Rating'] = fake.pyfloat(right_digits=2, positive=True, max_value=100)
        b_user[i]['Roughness(%)'] = fake.pyfloat(right_digits=2, min_value=0, max_value=1)
        b_user[i]['Resistance(g)'] = fake.pyint(max_value=5000)
        b_user[i]['Price(Target)'] = ((0.1*b_user[i]['Width'] + 0.2*b_user[i]['Height']) +
                                        (0.1*b_user[i]['Width'] + 0.2*b_user[i]['Height'])*0.002*b_user[i]['Amount'] +
                                        500*b_user[i]['Color'] +
                                        2.1*b_user[i]['Avg_speed'] + 
                                        1.5*b_user[i]['Weight'] +
                                        1.2*b_user[i]['Temperature(max)'] +
                                        16*b_user[i]['Rating'] +
                                        300*b_user[i]['Roughness(%)'] + 
                                        0.1*b_user[i]['Resistance(g)']) / 810


    return b_user


df = pd.DataFrame.from_dict(create_data(15000), orient='index')
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

### Define different nn models for different combinations of learning rate(0.1, 0.01, 0.001), optimizer(adam, sgd, rmsprop) and number hidden layers (1, 5, 25)

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
```

```python
def create_model(num_hiden, optim, l_rate):
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(10,)))
    for i in range(num_hiden):
        model.add(Dense(8, activation='relu'))
    model.add(Dense(1))

    if optim == 'adam':
        optimizer = Adam(learning_rate=l_rate)
    if optim == 'sgd':
        optimizer = SGD(learning_rate=l_rate)
    if optim == 'rms':
        optimizer = RMSprop(learning_rate=l_rate)

    model.compile(loss='mean_squared_error',
              optimizer=optimizer)

    return model
```

```python
import matplotlib.pyplot as plt
```

```python
alphas = [0.1, 0.01, 0.001]
optimizers = ['sgd', 'adam', 'rms']
nums_hiden = [1, 5, 25]
models = {}
histories = {}
for alpha in alphas:
    for optimizer in optimizers:
        for num_hiden in nums_hiden:
            models[(alpha, optimizer, num_hiden)] = create_model(num_hiden=num_hiden, optim=optimizer, l_rate=alpha)
            #print(models[(alpha, optimizer, num_hiden)].summary())
            histories[(alpha, optimizer, num_hiden)] = models[(alpha, optimizer, num_hiden)].fit(df_X.values, df_Y.values,epochs=15, validation_split=0.2)
```

```python
len(models)
```

### Plot learning curves for different combinations

```python
import matplotlib.pyplot as plt
```

```python
for alpha in alphas:
    for optimizer in optimizers:
        for num_hiden in nums_hiden:
            plt.plot(histories[(alpha, optimizer, num_hiden)].history['loss'])
            plt.plot(histories[(alpha, optimizer, num_hiden)].history['val_loss'])
            plt.title('model loss for parameters: alpha-{}, optimizer-{}, hidden layers-{}'.format(alpha, optimizer, num_hiden))
            plt.ylabel('loss')
            plt.xlabel('epoch')
            plt.legend(['train', 'test'], loc='upper left')
            plt.show()
```

### Compare DNN results

```python
best_dict = {}
best_dict['alpha'] = []
best_dict['optimizer'] = []
best_dict['num_h_layers'] = []
best_dict['train_loss'] = []
best_dict['test_loss'] = []
```

```python
for alpha in alphas:
    for optimizer in optimizers:
        for num_hiden in nums_hiden:
            best_dict['alpha'].append(alpha)
            best_dict['optimizer'].append(optimizer)
            best_dict['num_h_layers'].append(num_hiden)
            best_dict['train_loss'].append(histories[(alpha, optimizer, num_hiden)].history['loss'][-1])
            best_dict['test_loss'].append(histories[(alpha, optimizer, num_hiden)].history['val_loss'][-1])
```
